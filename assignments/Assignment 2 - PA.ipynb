### For Friday: Assignment 2 – upload a notebook to canvas that:

1. Defines a function that takes as its input a wavelength in meters and a temperature in Kelvin and outputs the emitted blackbody flux E∗λ in units of Wm−2m−1 as defined in Eq. 2.13 in Stull Chapter 2 p. 36
2. Uses that function to reproduce Stull figure 2.9
3. Defines a second function that outputs the blackbody radiance L∗λ in Wm−2m−1sr−1 using the equation at the bottom of the Finding the flux given the radiance notes
4. Makes another version of figure 2.9 that shows radiance instead of flux.

```python
def emittance(wavelength,temp):
    c1 = 3.74*10E7 
    c2= 1.44*10E3
    numer = c1/((wavelength)**5)
    ex = c2/(wavelength*temp)
    denom = math.exp(ex)- 1
    return numer/denom
```

```python
x_axis = np.linspace(0.1, 40,10000)
y_axis = [emittance(x,255) for x in x_axis]
fig,ax = plt.subplots()
ax.plot(x_axis,y_axis)
ax.set(xlabel='Wavelength(μm)', ylabel='Eλ* (W m-2 μm-1)',
       title='Planck radiant emittance at 255K')

plt.show()
```

```python
y = x_axis/np.pi
fig,ax = plt.subplots()
ax.plot(x_axis,y)
ax.set(xlabel='Wavelength(μm)', ylabel='Lλ* (W m-2 μm-1 sr-1)',
       title='Surface emitted radiance at 255K')

plt.show()
```
